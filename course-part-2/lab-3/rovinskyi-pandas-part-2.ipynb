```python
import pandas as pd
import numpy as np
```

```python
# Завантажити інформацію з файлу olympics.csv i почати зчитування інформації з першого рядка
df = pd.read_csv('olympics.csv', index_col = 0, skiprows=1)
df.head()
```

```python
# Вивести дані про таблицю
df.info()
```

```python
# Якого типу стовпці?
df.dtypes
```

```python
# Замінити тип стовпця на тип дати (стовпчик додано вручну)
df['unix_dates'] = np.random.randint(0, 1670161235, len(df))
df['normal_dates'] = pd.to_datetime(df['unix_dates'], unit='s')
df
```

```python
# Створити новий рядок таблиці, задавши значення тільки код та назва країни
df.loc['North Korea (NK)'] = np.nan
df
```

```python
# Вивести дані про таблицю. Відзначивши, що не всі поля заповненні
df.info(show_counts=True)
```

```python
# Заповнимо порожні – невизначені позиції нулем
df.fillna(0,inplace=True)
df
```

```python
# Відфільтрувати дані, показавши тільки country_5 (country_5 замінено Switzerland)
df[df.index.str.startswith("Switzerland")]
```

```python
# Відфільтрувати дані, показавши тільки country_5 (country_5 замінено Switzerland) та всі країни крім азійських
asia_countries = ('China','India','Indonesia','Pakistan','Bangladesh','Japan','Philippines','Vietnam','Turkey',
                  'Iran','Thailand','Myanmar','South Korea','Iraq','Afghanistan','Saudi Arabia','Uzbekistan',
                  'Malaysia','Yemen','Nepal','North Korea','Sri Lanka','Kazakhstan','Syria','Cambodia','Jordan',
                  'Azerbaijan','United Arab Emirates','Tajikistan','Israel','Laos','Lebanon','Kyrgyzstan',
                  'Turkmenistan', 'Singapore','Oman','State of Palestine','Kuwait','Georgia','Mongolia','Armenia',
                  'Qatar','Bahrain','Timor-Leste','Cyprus','Bhutan','Maldives','Brunei','Hong Kong','Taiwan', 
                  'Macao')

df[~df.index.str.startswith(asia_countries)]
```

```python
# Змінити назву стовпців 01 !, 02 !, 03 ! на Gold, Silver, Bronze

df.rename(columns={"01 !": "Gold", "02 !": "Silver", "03 !": "Bronze"}, inplace=True)
df
```

```python
# Створити нову таблицю тільки з трьома стовпцями Gold, Silver і Bronze

df_copy=df[['Gold', 'Silver', 'Bronze']]
df_copy
```

```python
# Знищити рядок Totals
df_copy = df_copy.drop(index='Totals')
df_copy
```

```python
# Вивести країни які мають золоті медалі
df_copy[df_copy['Gold']>0]
```

```python
# Підрахувати їх кількість
len(df_copy[df_copy['Gold']>0])
```

```python
# Яка країна має найбільшу кількість золотих медалей?
df_copy[df_copy['Gold'] == df_copy['Gold'].max()]
```

```python
# Які країни мають і золоті і срібні медалі?
df_copy[(df_copy['Gold']>0) & (df_copy['Silver']>0)]
```
