```python
# Дані знаходяться у файлі KN-4.xlsx 
import pandas as pd

df_kn4 = pd.read_excel('KN-4.xlsx')
df_kn4
```

```python
# Вивести розмір таблиці
df_kn4.shape
```

```python
# Вивести назви стопвців
df_kn4.columns
```

```python
# Вивести рядок Ваших даних
df_kn4[df_kn4['Student'] == "Ровінський Юрій"]
```

```python
# Невизначені значення NaN заповнити нулями
df_kn4.fillna(0, inplace=True)
df_kn4
```

```python
# Бали за тест перевести у 12-бальну систему оцінювання (помножити на 12, поділити на 100)
df_kn4['Test'] = df_kn4['Test']*12/100
df_kn4
```

```python
# Зарахувати по 2 бали за + за лабораторну та 1 бал за /
df_kn4['Lab1'] = df_kn4['Lab1'].replace(['+', '/'], [2, 1])
df_kn4['Lab2'] = df_kn4['Lab2'].replace(['+', '/'], [2, 1])
df_kn4
```

```python
# Створити стовпчик L7 на основі першої оцінки зі стовпця Lab-7. Дані перевести в числовий тип
tmp = df_kn4['Lab7'].str.split('/')
df_kn4['L7'] = tmp.apply(lambda x: int(x[0]))
df_kn4
```

```python
# Порахувати суму за Lab1,Lab2,L7,Test
df_kn4['Sum'] = df_kn4['Lab1'] + df_kn4['Lab2'] + df_kn4['Test'] + df_kn4['L7']
df_kn4
```

```python
# Перейменувати стовпчик L7
df_kn4=df_kn4.rename(columns={'L7':'L8'})
df_kn4
```

```python
# Яка найбільша сума?
df_kn4['Sum'].max()
```

```python
# Який номер рядка,що має найбільшу суму?
df_kn4[['Sum']].idxmax()
```

```python
# Вивести прізвище та ім’я студента, що має найбільшу суму
df_kn4[['Student']][df_kn4['Sum']==df_kn4['Sum'].max()]
```

```python
# Вивести дані про студентів, що отримали сумарний нуль
df_kn4[df_kn4['Sum']==0]
```

```python
# Хто має більше ніж 15 сумарних балів?
df_kn4[df_kn4['Sum']>15]
```

```python
# Скільки студентів мають більше 15 балів?
len(df_kn4[df_kn4['Sum']>15])
```

```python
# Посортувати по стовпцю суми
df_kn4.sort_values(by=['Sum'])
```

```python
# Хто здав першу лабораторну?
df_kn4[['Student', 'Lab1']][df_kn4['Lab1'].isin([2,1])]
```

```python
# Додати стовпчик для лаб4 зі значенням 2
df_kn4['Lab4'] = 2
df_kn4
```

```python
# Вивести прізвища і суми
tmp = df_kn4['Student'].str.split(' ')
df_kn4['Surname'] = tmp.apply(lambda student_data: student_data[0])
df_kn4[['Surname', 'Sum']]
```

```python
# Створити стовпчик допуск, де задати «dopusk» якщо сума балів більша 15
df_kn4 = df_kn4.drop('Surname', axis=1)
df_kn4['Dopusk'] = df_kn4['Sum'].apply(lambda final_sum: "dopusk" if final_sum > 15 else "nedopusk")
df_kn4
```

```python
# Показати тих, хто має допуск
df_kn4[['Student', 'Dopusk']][df_kn4['Dopusk'] == "dopusk"]
```
