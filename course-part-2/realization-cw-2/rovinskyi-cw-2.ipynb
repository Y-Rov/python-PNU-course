## Дослідження кількості світового населення за країнами (1970-2050рр.)

### Обрані датасети:
1. https://www.kaggle.com/datasets/tusharaggarwal27/2022-dec-world-populations
2. https://www.kaggle.com/datasets/iamsouravbanerjee/world-population-dataset

```python
import pandas as pd
import numpy as np
```

```python
first_dataframe = pd.read_csv("world_population_1.csv")
first_dataframe
```

```python
second_dataframe = pd.read_csv("world_population_2.csv")
second_dataframe
```

```python
second_dataframe[['pop2030', 'pop2050']] *= 1000
result_dataframe = pd.merge(first_dataframe, second_dataframe[['country', 'pop2030', 'pop2050']], 
'left', left_on='Country/Territory', right_on='country')
result_dataframe = result_dataframe.astype({ 'pop2030': np.int32, 'pop2050': np.int32 })
result_dataframe.rename(columns={"pop2030": "2030 Predicted population",
 "pop2050": "2050 Predicted population"}, inplace=True)
result_dataframe.drop(columns=['CCA3', 'Capital', 'country'])
result_dataframe = result_dataframe[['Rank', 'Country/Territory', 'Continent', '1970 Population', 
'1980 Population', '1990 Population', '2000 Population', '2010 Population', '2015 Population', 
'2020 Population', '2022 Population', '2030 Predicted population', '2050 Predicted population',
'Area (km²)', 'Density (per km²)', 'Growth Rate', 'World Population Percentage']]
result_dataframe
```
