## Дослідження динаміки зміни кількості світового населення за країнами (1970-2050рр.)

### Обрані датасети:
https://www.kaggle.com/datasets/tusharaggarwal27/2022-dec-world-populations
https://www.kaggle.com/datasets/iamsouravbanerjee/world-population-dataset

```python
# Імпорт необхідних бібліотек
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as tckr
```

### Отримання даних:

```python
# Dataset №1
first_df = pd.read_csv("world_population_1.csv")
first_df
```

```python
# Dataset №2
second_df = pd.read_csv("world_population_2.csv")
second_df
```

### Підготовка даних для подальшої роботи

```python
# Домноження стовпців pop2030 і pop2050 на 1000 - приведення до звичайного числа.
# Попередні дані були виражені в тисячах, тобто 1.45 = 1450.
transformed_second_df = second_df.apply(lambda value: value * 1000 if value.name in ['pop2030', 'pop2050'] else value) 

# Об'єднання двох датасетів у один
result_df = pd.merge(first_df, transformed_second_df[['country', 'pop2030', 'pop2050']], 
'left', left_on='Country/Territory', right_on='country')

# Приведення типу стовпців pop2030 і pop2050 до цілого числа
result_df = result_df.astype({ 'pop2030': np.int32, 'pop2050': np.int32 })

# Перейменування назв деяких стовпців
result_df.rename(
    columns= {
        "pop2030": "2030 Predicted population", "pop2050": "2050 Predicted population",
        "World Population Percentage": "World Population (%)"},
    inplace=True)

# Вилучення зайвої інформації
result_df.drop(columns=['CCA3', 'Capital', 'country'])

# Визначення наступного порядку для стовпців
result_df = result_df[['Rank', 'Country/Territory', 'Continent', '1970 Population', '1980 Population',
                       '1990 Population', '2000 Population', '2010 Population', '2015 Population', 
                       '2020 Population', '2022 Population', '2030 Predicted population',
                       '2050 Predicted population', 'Area (km²)', 'Density (per km²)',
                       'Growth Rate', 'World Population (%)']]
result_df
```

### Дослідження

#### Створення обчислювальних полів

```python
# Передбачений темп зростання кількості населення (< 1 - спад, > 1 - ріст)
# 2030 рік
result_df['2030 Predicted growth rate'] = result_df['2030 Predicted population'] / result_df['2022 Population']

# 2050 рік
result_df['2050 Predicted growth rate'] = result_df['2050 Predicted population'] / result_df['2022 Population']

# Переважання народжуваності чи смертності у майбутньому (2030-2050рр.)
result_df['Future dominance'] = (result_df['2050 Predicted population'] / result_df['2030 Predicted population']).apply(lambda value: 'Birth' if value > 1 else 'Death')
result_df
```

#### Групування

```python
# За континентами + обчислення теперішньої кількості світового населення
# Примітка - станом на жовтень 2022 року (останнє оновлення основного датасету)
population_amount = result_df[['Continent', '2022 Population']]
population_amount = population_amount.groupby('Continent').sum()
population_amount.loc['Total:'] = population_amount['2022 Population'].sum()
population_amount
```

```python
# За майбутніми тенденціями + обчислення кількості країн з цими тенденціями
future_dominance = result_df[['Country/Territory', 'Future dominance']]
future_dominance = future_dominance.groupby('Future dominance').count()
future_dominance
```

```python
# За континентами + обчислення сумарної густоти населення разом із площею континенту
population_density = result_df[['Continent', 'Density (per km²)', 'Area (km²)']]
population_density = population_density.groupby('Continent').sum()
population_density.loc['Total:'] = [population_density['Density (per km²)'].sum(),
                                    population_density['Area (km²)'].sum()]
population_density = population_density.astype({'Area (km²)' : np.int64})
population_density
```

#### Маски

```python
# Дізнатися країни, у яких кількість населення, станом на зараз, більша за 50 мільйонів
massive_countries = result_df[['Country/Territory', '2022 Population']][result_df['2022 Population'] > 50000000]

# Сортування по зростанню кількості
massive_countries.sort_values('2022 Population')
```

```python
# Дізнатися в яких країнах передбачений приріст населення (2050 рік)
countries = result_df[['Country/Territory','2050 Predicted growth rate']][result_df['2050 Predicted growth rate'] > 1]

# Вивід перших десяти країн
countries.iloc[:10]
```

```python
# Вивести країни, де частка світового населення складає менше 0.05%
small_countries = result_df[['Country/Territory','World Population (%)']][(result_df['World Population (%)'] < 0.05)]
small_countries
```

#### Визначення індексу

```python
# Новий індекс - місце у рейтингу ('Rank')
result_df_with_rank_index = result_df.set_index('Rank')
result_df_with_rank_index.sort_index(inplace=True)
result_df_with_rank_index
```
